Load in libraries

```julia
using PyPlot
using DataFrames
```

Load in data

```julia
trial_dat = readtable("trial_by_trial_values_initalise0_appended.csv")
```

```julia
# select AB subs 
trial_dat_AB = trial_dat[find(trial_dat[:order_condition].==1),:];

#list of these subjects
AB_subs = unique(trial_dat_AB[:sub])

dat_compile_A = ones(length(AB_subs),300)*NaN
dat_compile_B = ones(length(AB_subs),300)*NaN

dat_compile_A_arithmetic = ones(length(AB_subs),300)*NaN
dat_compile_B_arithmetic = ones(length(AB_subs),300)*NaN

for i = 1:length(AB_subs)
    
    #print(length(find((trial_dat_AB[:block].==0) .& (trial_dat_AB[:sub].==AB_subs[i]))))
    #print(" ")
    
    n_trials_A = length(trial_dat_AB[find((trial_dat_AB[:block].==0) .& (trial_dat_AB[:sub].==AB_subs[i])),:avreward_estimate])
    
    for x = 1:n_trials_A
    
        dat_compile_A[i,x] = trial_dat_AB[find((trial_dat_AB[:block].==0) .& (trial_dat_AB[:sub].==AB_subs[i])),:avreward_estimate][x]
        dat_compile_A_arithmetic[i,x] = trial_dat_AB[find((trial_dat_AB[:block].==0) .& (trial_dat_AB[:sub].==AB_subs[i])),:avreward_arithmetic][x]
        
    end
    
    n_trials_B = length(trial_dat_AB[find((trial_dat_AB[:block].==1) .& (trial_dat_AB[:sub].==AB_subs[i])),:avreward_estimate])

    for x = 1:n_trials_B
    
        dat_compile_B[i,x] = trial_dat_AB[find((trial_dat_AB[:block].==1) .& (trial_dat_AB[:sub].==AB_subs[i])),:avreward_estimate][x]
        dat_compile_B_arithmetic[i,x] = trial_dat_AB[find((trial_dat_AB[:block].==1) .& (trial_dat_AB[:sub].==AB_subs[i])),:avreward_arithmetic][x]

    end
    
    
end
```

```julia
Q_means_ABsubs_A = mean(dat_compile_A,1)
Q_means_ABsubs_A = Q_means_ABsubs_A[find(!isnan(Q_means_ABsubs_A))]

Q_means_ABsubs_A_arithmetic = mean(dat_compile_A_arithmetic,1)
Q_means_ABsubs_A_arithmetic = Q_means_ABsubs_A_arithmetic[find(!isnan(Q_means_ABsubs_A_arithmetic))]

plot(Q_means_ABsubs_A, "*", markersize = 4)
plot(Q_means_ABsubs_A_arithmetic, "*", markersize = 4, color = "g")

xlabel("Trial n")
ylabel("Q estimate (average over subs)")
ylim([0, 20])
xlim([0,160])
title("AB subs, block A")
```

```julia
Q_means_ABsubs_B = mean(dat_compile_B,1)
Q_means_ABsubs_B = Q_means_ABsubs_B[find(!isnan(Q_means_ABsubs_B))]

Q_means_ABsubs_B_arithmetic = mean(dat_compile_B_arithmetic,1)
Q_means_ABsubs_B_arithmetic = Q_means_ABsubs_B_arithmetic[find(!isnan(Q_means_ABsubs_B_arithmetic))]

plot(Q_means_ABsubs_B, "*", markersize = 4)
plot(Q_means_ABsubs_B_arithmetic, "*", markersize = 4, color = "g")

xlabel("Trial n")
ylabel("Q estimate (average over subs)")
ylim([0, 20])
xlim([0,140])
title("AB subs, block B")
```
