Load in libraries

```julia
using PyPlot
using DataFrames
```

Load in data

```julia
rt_dat = readtable("rt_repeated_measures.csv")
```

```julia
y_means = [mean(filter(!isnan,rt_dat[:1])), 
    mean(filter(!isnan,rt_dat[:2])), 
    mean(filter(!isnan,rt_dat[:3])), 
    mean(filter(!isnan,rt_dat[:4]))]
```

```julia
error = [std(filter(!isnan,rt_dat[:1]))/sqrt(length(filter(!isnan,rt_dat[:1]))), 
    std(filter(!isnan,rt_dat[:2]))/sqrt(length(filter(!isnan,rt_dat[:2]))),  
    std(filter(!isnan,rt_dat[:3]))/sqrt(length(filter(!isnan,rt_dat[:3]))),  
    std(filter(!isnan,rt_dat[:4]))/sqrt(length(filter(!isnan,rt_dat[:4])))]
```

```julia
y_means+error
```

```julia
?bar
```

```julia
#colours = ["#EC602A", "#C99262", "#9A8DAD", "#54B143"]
colours = "grey"
x = [1,2,3,4]

my_ticks = ["accept_lowDV", "accept_highDV", "reject_lowDV", "reject_highDV"]
bar(my_ticks, y_means, color=colours, edgecolor = ["cyan", "cyan","cyan","cyan"], yerr=error, capsize=10, ecolor = "red", linewidth = 1.5)

ylabel("mean rt (z_trans)")

ylim([0,0.75])
annotate("n.s (p=0.09)",
xy=[1.2;0.7],
fontsize=8)

plot([1, 2], [0.65, 0.65], lw=1, "--", color="black")

annotate("*",
xy=[3.5;0.65],
fontsize=30)

plot([3, 4], [0.65, 0.65], lw=1, "--", color="black")
```
